```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
data= {"Areas"
        :[1420, 1950, 2600, 1780, 2120, 1350, 2830, 1640, 1240, 2480, 
                  2060, 1390, 2920, 1750, 2280, 1970, 1340, 2460, 2700, 1550, 
                  1830, 2240, 1180, 2950, 1630, 2100, 2550, 1440, 1970, 2660, 
                  2220, 1320, 1880, 1510, 2750, 2160, 1280, 2030, 2490, 2810, 
                  1580, 1920, 2670, 2290, 1130, 2850, 1520, 2100, 2610],
       "Prices":[220000, 310000, 420000, 290000, 350000, 200000, 480000, 
                   270000, 180000, 400000, 320000, 240000, 500000, 260000, 
                   380000, 330000, 210000, 410000, 460000, 250000, 310000, 
                   370000, 170000, 490000, 290000, 350000, 430000, 240000, 
                   320000, 440000, 360000, 190000, 330000, 260000, 450000, 
                   370000, 200000, 320000, 410000, 480000, 270000, 330000, 
                   450000, 370000, 160000, 490000, 250000, 350000, 430000]}
df=pd.DataFrame(data)
df.head(5)
```

```python
plt.xlabel("Areas")
plt.ylabel("Prices")
plt.scatter(df["Areas"],df.Prices)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["Areas"]],df.Prices)
```

```python
reg.predict([[2200]])
```

```python
coef=reg.coef_
coef
```

```python
reg.intercept_
```

```python
#y=mx+c
coef*2200+reg.intercept_
#the value is equal to the valuepref=dicted
```

```python
plt.xlabel("Areas")
plt.ylabel("Prices")
plt.scatter(df["Areas"],df.Prices)
plt.plot(df["Areas"],reg.predict(df[["Areas"]]),color="yellow")
```
